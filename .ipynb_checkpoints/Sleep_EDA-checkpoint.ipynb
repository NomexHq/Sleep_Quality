```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import missingno as msno
from statsmodels.formula.api import ols
import statsmodels.api as sm
import statsmodels.stats.api as sms
import scipy
from scipy import stats
from scipy.stats import norm
from scipy.stats import ttest_1samp
%matplotlib inline
```

```python
df_full = pd.read_csv('df_full.csv')
```

# EDA & Visualisations

```python
# a quick visualisation of corr and distributions

sns.pairplot(df_full);
```

```python
# looking at correlation between vaiables

round(df_full.corr(), 2)
```

```python
# which ones are highly correlated?

abs(df_full.corr()) > 0.75
```

```python
# corr heatmap

sns.heatmap(df_full.corr(), center=0); 
# high correlation with sleep_duration becuase it is used to calculate the coefficient; 
# there is also correlation with start time
# other variables show little to no correlation
```

### Sleeping_quality variable

```python
# histogram of sleep quality index

plt.hist(df_full.Sleep_quality, bins=20)
plt.xlabel('Sleep quality')
plt.ylabel('Number of occurances')
plt.show()

# skewed to the right with potential outliers on the left
```

```python
# sleep quality PMF

pmf_sleep = df_full.Sleep_quality.value_counts().sort_index() / len(df_full.Sleep_quality)
pmf_sleep.plot(kind='bar');
plt.xlabel('Index')
plt.ylabel('PMF')
```

```python
# Sleep quality KDE plot 

sns.kdeplot(df_full.Sleep_quality);
```

```python
# sleep quality boxplot

sns.set(style="whitegrid")
sns.boxplot(x=df_full.Sleep_quality);

# allows to see two outliers 
```

```python
# removing the  outliers

# outliers = df_full.Sleep_quality.value_counts().sort_index()
# outliers

no_out = df_full.drop(df_full.index[df_full['Sleep_quality'] == 27], inplace = True)
```

### Going_to_bed variable

```python
# histogram of going to bed hours

plt.hist(df_full.Start_time, bins=20)
plt.xlabel('Times')
plt.ylabel('Number of occurances')
plt.show()

# decided not to treat as a categorical variable
```

```python
# going to bed times PMF

pmf_times = df_full.Start_time.value_counts().sort_index() / len(df_full.Start_time)
pmf_times.plot(kind='bar');
plt.xlabel('Times')
plt.ylabel('PMF')
```

```python
# going to bed times boxplot

sns.set(style="whitegrid")
sns.boxplot(x=df_full.Start_time);
```

```python
# number of steps boxplot

sns.set(style="whitegrid")
sns.boxplot(x=df_full.Activity_steps);

# shows an outlier
```

### Hours spent in bed

```python
# histogram of number of hours spent in bed

plt.hist(df_full.Time_in_bed, bins=20, histtype='step') # changing the histtype for fun
plt.xlabel('Minutes')
plt.ylabel('Number of occurances')
plt.show()

# approx normal
```

### Quality of sleep by day of week

```python
sns.barplot(x=df_full['Sleep_quality'], y=df_full['Start_Dayofweek'], data=df_full).set_title("Quality of sleep by day of week");
plt.savefig('Quality of sleep by day of week')

# overall, confirms intutition but Tuesday is a surprise. 
# Also, always thought that my Sunday sleep was bad, but data showed otherwise
```
